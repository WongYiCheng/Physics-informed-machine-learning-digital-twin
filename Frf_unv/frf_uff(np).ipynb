# Import Modules

```python
import numpy as np
import pyuff
import os
save_to_folder = 'frf_unv'
class_name = 'UD_T35_R1'

# all_frf_data = np.load(f'450,60,2 Dataset\\Frf_2\\frf_(N){class_name}_T00_R1.npy')
all_frf_data = np.load(rf'D:\UM Document\FYP_ML\450,60,2 Dataset\Frf_3\frf_(T){class_name}.npy')
trial_to_pick = 19 
frf_data = all_frf_data[trial_to_pick, :, : 401] # Resulting shape: (9, 401)
print(frf_data.shape)
```

# Dataset 164

```python
dataset164 = pyuff.prepare_164(
    units_code=1,
    units_description='SI units',
    temp_mode=1,
    length=1.0,
    force=1.0,
    temp=1.0,
    temp_offset=1.0)

if save_to_folder:
    if not os.path.exists(save_to_folder):
        os.makedirs(save_to_folder)
    uffwrite = pyuff.UFF(os.path.join(save_to_folder, f'Frf_{class_name}.unv'))
    uffwrite._write_set(dataset164, 'add')
```

```python
uff_datasets = []
# frequency = np.linspace(0, 1024, 2049) # for 1024Hz sampling rate and 2049 points
frequency = np.linspace(0, 200, 401) # for 200Hz sampling rate and 401 points


# Loop through the 9 sensors 
for sensor_idx in range(frf_data.shape[0]):
    
    # Extract the complex FRF for this sensor
    frf_vector = frf_data[sensor_idx, :]
    
    response_node = sensor_idx + 1
    reference_node = 10
    
    print(f'Adding Sensor {response_node} from Trial {trial_to_pick + 1}')

    data58 = pyuff.prepare_58(
        func_type=4,
        rsp_node=response_node,
        rsp_dir=3,
        ref_node=reference_node,
        ref_dir=3,
        data=frf_vector,
        x=frequency,
        id1=f'frf_{trial_to_pick + 1}',
        # id4=f'Sensor_{response_node}',
        rsp_ent_name=f'{response_node}Z',
        ref_ent_name='10Z',
        abscissa_spacing=1,
        abscissa_spec_data_type=18,
        ordinate_spec_data_type=12,
        orddenom_spec_data_type=13,
        abscissa_axis_units_lab='Hz',
        ordinate_axis_units_lab='m/s2',
        orddenom_axis_units_lab='N'
    )

    # Write to file
    uffwrite._write_set(data58, mode='add')

print(f'[{class_name}] Successfully wrote FRF data to UNV file.')
```
